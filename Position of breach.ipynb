```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = "Times New Roman"
figure(figsize=(8, 6), dpi=80)
```

```python
# velocity
df1=pd.read_excel('data/position of breach/Velocity1.xlsx')
```

```python
df1.head(500)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df1["Station (m)"],df1["A"],label="Position Left Max Velocity")
plt.plot(df1["Station (m)"],df1["B"],label="Position Right Max Velocity")
# plt.plot(df1["Station (m)"],df1[""])
plt.plot(df1["Station (m)"],df1["C"],label="Position Center Max Velocity")
# grid
plt.grid(b=True)
# label
plt.title("Velocity on 'River_Centerline' ",fontweight="bold")
plt.ylabel('Velocity Max (m/s)')
plt.xlabel('Station[m]')
# legend
plt.legend(loc='best')
plt.savefig('final/position of breach/velocity.jpg',dpi=300)
```

```python
# wse
df2=pd.read_excel('data/position of breach/WSE1.xlsx')
df2.head()
```

```python
df2.head(500)
```

```python
df2 =df2.interpolate()
plt.figure(figsize=(10,6))
plt.plot(df2["Station (m)"],df2["A"],label="Position Left Max WSE")
plt.plot(df2["Station (m)"],df2["B"],label="Position Right Max WSE")
# plt.plot(df1["Station (m)"],df1[""])
plt.plot(df2["Station (m)"],df2["C"],label="Position Center Max WSE")
plt.plot(df2["Station (m)"],df2["D"],label="Terrain Profile")

# grid
plt.grid(b=True)
# label
plt.title("Water Surface Elevation 'River_Centerline' ",fontweight="bold")
plt.ylabel('Water Surface Elevation(meters)')
plt.xlabel('Station[m]')
# legend
plt.legend(loc='best')
plt.savefig('final/position of breach/wse.jpg',dpi=300)
```

```python
# section 1
#time vs all parameters
# fullview
plt.figure(figsize=(10,6))

plt.plot(df1["Station (m)"],df1["A"],label="Position Left Flow")
plt.plot(df1["Station (m)"],df1["B"],label="Position Right Flow")
# plt.plot(df1["Station (m)"],df1[""])
plt.plot(df1["Station (m)"],df1["C"],label="Position Center Flow")



# grid
plt.grid(b=True)
# label
plt.title("Peak Discharge for Different Values of Dam Breach Position at Location 1 ",fontweight="bold")
plt.ylabel('Flow(CMS)')
plt.xlabel('Date/Time')

# legend
plt.legend(loc='best')

plt.savefig('final/position of breach/Sheets/Peakdischarge1.jpg',dpi=300)
# for max value
maxValues = df1.max()
 
print(maxValues)
```
