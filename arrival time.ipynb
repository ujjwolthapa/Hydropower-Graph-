```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(8, 6), dpi=80)
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = "Times New Roman"
```

```python
df1=pd.read_excel('data/Breachformationtime/arrivaltime.xlsx')
```

```python
df1.plot()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df1["Station (m)"],df1["Breach Formation Time 0.1 Arrival Time (hours)"],label="Breach Formation Time 0.1")
plt.plot(df1["Station (m)"],df1["Breach Formation Time 0.15 Arrival Time (hours)"],label="Breach Formation Time 0.15")
# plt.plot(df1["Station (m)"],df1[""])
plt.plot(df1["Station (m)"],df1["Breach Formation Time 0.2 Arrival Time (hours)"],label="Breach Formation Time 0.2")
plt.plot(df1["Station (m)"],df1["Breach Formation Time 0.25 Arrival Time (hours)"],label="Breach Formation Time 0.25")
plt.plot(df1["Station (m)"],df1["Breach Formation Time 0.3 Arrival Time (hours)"],label="Breach Formation Time 0.3")
# grid
plt.grid(b=True)
# label
plt.title("Arrival Time on 'River_Centerline' ",fontweight="bold")
plt.ylabel('Arrival Time (Hours)')
plt.xlabel('Station[m]')
# legend
plt.legend(loc='lower right')
plt.savefig('final/Breachformationtime/arrivaltime.jpg',dpi=300)
```
