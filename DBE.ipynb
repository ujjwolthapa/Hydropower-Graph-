```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = "Times New Roman"
figure(figsize=(8, 6), dpi=80)
```

```python
# velocity
df1=pd.read_excel('data/DBE/Velocity1.xlsx')
```

```python
df1.head()
```

```python
df1.plot()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df1["Station (m)"],df1["A"],label="Dam Breach Elevation 1480 Max Velocity")
plt.plot(df1["Station (m)"],df1["B"],label="Dam Breach Elevation 1490 Max Velocity")
# plt.plot(df1["Station (m)"],df1[""])
plt.plot(df1["Station (m)"],df1["C"],label="Dam Breach Elevation 1500 Max Velocity")
plt.plot(df1["Station (m)"],df1["D"],label="Dam Breach Elevation 1510 Max Velocity")
plt.plot(df1["Station (m)"],df1["E"],label="Dam Breach Elevation 1520 Max Velocity")
# grid
plt.grid(b=True)
# label
plt.title("Velocity on 'River_Centerline' ",fontweight="bold")
plt.ylabel('Velocity Max (m/s)')
plt.xlabel('Station[m]')
# legend
plt.legend(loc='best')
plt.savefig('final/DBE/velocity.jpg',dpi=300)
```

```python
# wse
df2=pd.read_excel('data/DBE/WSE1.xlsx')
df2.head()
```

```python
df2=df2.interpolate()
plt.figure(figsize=(10,6))
plt.plot(df2["Station (m)"],df2["A"],label="Dam Breach Elevation 1480 Max WSE")
plt.plot(df2["Station (m)"],df2["B"],label="Dam Breach Elevation 1490 Max WSE")
# plt.plot(df1["Station (m)"],df1[""])
plt.plot(df2["Station (m)"],df2["C"],label="Dam Breach Elevation 1500 Max WSE")
plt.plot(df2["Station (m)"],df2["D"],label="Dam Breach Elevation 1510 Max WSE")
plt.plot(df2["Station (m)"],df2["E"],label="Dam Breach Elevation 1520 Max WSE")
plt.plot(df2["Station (m)"],df2["F"],label="Terrain Profile")
# grid
plt.grid(b=True)
# label
plt.title("Water Surface Elevation 'River_Centerline' ",fontweight="bold")
plt.ylabel('Water Surface Elevation(meters)')
plt.xlabel('Station[m]')
# legend
plt.legend(loc='best')
plt.savefig('final/DBE/wse.jpg',dpi=300)
```
