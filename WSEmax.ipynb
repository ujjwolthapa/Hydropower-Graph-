```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(8, 6), dpi=80)
```

```python
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = "Times New Roman"
```

```python
df1=pd.read_excel('data/Breachformationtime/WSEmax1.xlsx')
```

```python
df1.head()
```

```python
df1.head(500
        )
```

```python
df1= df1.interpolate()
```

```python
df1.plot()
```

```python
# without interpolate
plt.figure(figsize=(10,6))
plt.plot(df1["Station (m)"],df1["A"],label="Breach Formation Time 0.1 WSE Max")
plt.plot(df1["Station (m)"],df1["B"],label="Breach Formation Time 0.15 WSE Max")
# plt.plot(df1["Station (m)"],df1[""])
plt.plot(df1["Station (m)"],df1["C"],label="Breach Formation Time 0.2 WSE Max")
plt.plot(df1["Station (m)"],df1["D"],label="Breach Formation Time 0.25 WSE Max")
plt.plot(df1["Station (m)"],df1["E"],label="Breach Formation Time 0.3 WSE Max")
plt.plot(df1["Station (m)"],df1["F"],label="Terrain Profile")
# grid
plt.grid(b=True)
# label
plt.title("Water Surface Elevation 'River_Centerline' ",fontweight="bold")
plt.ylabel('Meters')
plt.xlabel('Station[m]')
# legend
plt.legend(loc='best')
plt.savefig('final/Breachformationtime/1wsemax.jpg',dpi=300)
```
