```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = "Times New Roman"
figure(figsize=(8, 6), dpi=80)
```

```python
df1=pd.read_excel('data/Breachformationtime/VelocityMax1.xlsx')
```

```python
# csfont = {'fontname':'Times New Roman'}
# # write your code related to basemap here
# # plt.title('title',**csfont)
```

```python
df1.plot()
```

```python
# template
plt.figure(figsize=(10,6))
plt.plot(df1["Station (m)"],df1[0.1],label="Breach Formation Time 0.1 Velocity Max")
plt.plot(df1["Station (m)"],df1[0.15],label="Breach Formation Time 0.15 Velocity Max")
# plt.plot(df1["Station (m)"],df1[""])
plt.plot(df1["Station (m)"],df1[0.2],label="Breach Formation Time 0.2 Velocity Max")
plt.plot(df1["Station (m)"],df1[0.25],label="Breach Formation Time 0.25 Velocity Max")
plt.plot(df1["Station (m)"],df1[0.3],label="Breach Formation Time 0.3 Velocity Max")
# grid
plt.grid(b=True)
# label
plt.title("Velocity on 'River_Centerline' ",fontweight="bold")
plt.ylabel('Velocity Max (m/s)')
plt.xlabel('Station[m]')
# legend
plt.legend(loc='best')
plt.savefig('final/Breachformationtime/velocitymax.jpg',dpi=300)
```
